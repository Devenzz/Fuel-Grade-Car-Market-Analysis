```python
import pandas as pd
import numpy as np

# Function to convert price to Lakh
def convert_price_to_lakh(price_str):
    if pd.isna(price_str):
        return np.nan
    price_str = price_str.replace(',', '').strip()

    # Handle range like "7.41 - 10 Lakh"
    if '-' in price_str:
        parts = price_str.split('-')
        values = [p.strip() for p in parts]
        # Take average of the range
        try:
            lower = float(values[0].replace('Lakh','').replace('Cr','').strip())
            upper = float(values[1].replace('Lakh','').replace('Cr','').strip())
        except:
            return np.nan
        # Convert to Lakh if in Cr
        if 'Cr' in values[0] or 'Cr' in values[1]:
            lower *= 100
            upper *= 100
        return (lower + upper)/2
    else:
        # Single value
        if 'Cr' in price_str:
            return float(price_str.replace('Cr','').strip())*100
        elif 'Lakh' in price_str:
            return float(price_str.replace('Lakh','').strip())
        else:
            try:
                return float(price_str)
            except:
                return np.nan

# Apply the function to entire column
df_cars['Price_in_Lakh'] = df_cars['price_ex_showroom'].apply(convert_price_to_lakh)

# View all rows
pd.set_option('display.max_rows', None)
print(df_cars[['price_ex_showroom', 'Price_in_Lakh']])
```
