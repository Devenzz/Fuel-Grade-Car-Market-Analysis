```python
import pandas as pd
import re

# --- Enhanced Car Type Classification Function ---
def assign_car_type(row):
    model = str(row["model"]).lower()
    brand = str(row["brand"]).lower()

    # ----- SUVs -----
    if any(x in model for x in [
        "suv", "x1", "x3", "x5", "x7", "gls", "gle", "glc", "gla", "glb",
        "fortuner", "harrier", "hector", "safari", "creta", "venue", "seltos",
        "scorpio", "bolero", "thar", "compass", "kodiaq", "tiguan", "gloster",
        "kiger", "magnite", "taigun", "kushaq", "sonet", "brezza", "fronx",
        "xuv", "venue", "alcazar", "selta", "nexon", "punch", "jimny", "gv80",
        "q3", "q5", "q7", "q8", "xc40", "xc60", "xc90", "land cruiser"
    ]) or brand in [
        "jeep", "land rover", "mahindra", "tata", "toyota", "mg", "skoda", "hyundai", "kia", "nissan"
    ]:
        return "SUV"

    # ----- Sedans -----
    elif any(x in model for x in [
        "c-class", "e-class", "s-class", "a-class", "verna", "city", "ciaz", "virtus",
        "slavia", "rapid", "camry", "octavia", "superb", "tigor", "dzire", "amaze",
        "aura", "elantra", "passat", "a4", "a6", "s60", "s90", "tigor", "indigo"
    ]) or brand in [
        "mercedes", "audi", "bmw", "volvo", "skoda", "honda", "toyota", "hyundai", "maruti", "vw", "volkswagen"
    ] and "suv" not in model:
        return "Sedan"

    # ----- Hatchbacks -----
    elif any(x in model for x in [
        "swift", "baleno", "i20", "i10", "altroz", "tiago", "ignis", "kwid", "celerio",
        "glanza", "getz", "jazz", "bolt", "indica", "micra", "go", "corsa", "polo", "nano"
    ]):
        return "Hatchback"

    # ----- Coupes / Sports Cars -----
    elif any(x in model for x in [
        "coupe", "m4", "m8", "z4", "amg", "gt", "mustang", "supra", "911", "718", "huracan",
        "urus", "f-type", "r8", "a110", "lc", "db11", "dbs", "vantage", "mc20", "488", "296", "roma"
    ]) or brand in [
        "ferrari", "lamborghini", "porsche", "aston martin", "mclaren"
    ]:
        return "Coupe"

    # ----- Convertibles -----
    elif any(x in model for x in ["roadster", "spyder", "convertible", "cabriolet"]):
        return "Convertible"

    # ----- Pickup / Cargo -----
    elif any(x in model for x in ["pickup", "cargo", "hilux", "isuzu", "d-max", "yodha", "camper"]) or brand in ["isuzu"]:
        return "Pickup / Cargo"

    # ----- MPV / MUV / Vans -----
    elif any(x in model for x in [
        "innova", "carens", "ertiga", "triber", "evalia", "marazzo", "tour", "hexa",
        "vellfire", "rumion", "alcazar", "xl6", "byd e6"
    ]):
        return "MPV / MUV"

    # ----- Luxury Sedans / Coupes -----
    elif brand in ["rolls-royce", "bentley", "maybach"]:
        return "Luxury Sedan / Coupe"

    # ----- Fallback -----
    else:
        # Try to guess based on brand if model not matched
        if brand in ["jeep", "mahindra", "tata", "toyota", "mg"]:
            return "SUV"
        elif brand in ["ferrari", "lamborghini", "porsche", "aston martin", "mclaren"]:
            return "Coupe"
        elif brand in ["maruti", "hyundai", "honda", "nissan", "tata", "vw", "volkswagen"]:
            return "Sedan"
        else:
            return "Sedan"

# --- Apply function ---
df_cars["car_type"] = df_cars.apply(assign_car_type, axis=1)

# --- Check results ---
print(df_cars["car_type"].value_counts())
print(df_cars[["brand", "model", "car_type"]].sample(20))
```
