```python
import pandas as pd
import numpy as np

# Assuming df_carss is your active dataframe

# Define common variant suffixes for Indian market
variant_suffixes = [
    "Base", "LXi", "VXi", "ZXi", "ZXi+", 
    "E", "EX", "S", "SX", "SX(O)", 
    "Trend", "Titanium", "Signature", "AX5", "AX7"
]

# Define base models for which well generate variants
base_variants = [
    ("Maruti Suzuki", "Swift", "Petrol", 25.75, 7.3, 1197, 5, "India", "Hatchback"),
    ("Maruti Suzuki", "Baleno", "Petrol", 22.9, 7.7, 1197, 5, "India", "Hatchback"),
    ("Maruti Suzuki", "Dzire", "Petrol", 25.7, 7.7, 1197, 5, "India", "Sedan"),
    ("Hyundai", "Creta", "Petrol", 18.0, 16.9, 1497, 5, "South Korea", "SUV"),
    ("Hyundai", "Venue", "Petrol", 18.5, 10.0, 1197, 5, "South Korea", "SUV"),
    ("Tata", "Nexon", "Petrol", 24.0, 10.6, 1497, 5, "India", "SUV"),
    ("Tata", "Altroz", "Petrol", 19.0, 8.4, 1497, 5, "India", "Hatchback"),
    ("Tata", "Harrier", "Diesel", 16.5, 19.6, 1956, 5, "India", "SUV"),
    ("Toyota", "Fortuner", "Diesel", 12.0, 41.0, 2755, 7, "Japan", "SUV"),
    ("Toyota", "Innova Crysta", "Diesel", 14.5, 22.0, 2393, 7, "Japan", "MPV / MUV"),
    ("Kia", "Seltos", "Petrol", 20.0, 15.8, 1497, 5, "South Korea", "SUV"),
    ("Kia", "Sonet", "Petrol", 19.5, 11.3, 1493, 5, "South Korea", "SUV"),
    ("Mahindra", "XUV700", "Diesel", 16.5, 20.5, 2184, 7, "India", "SUV"),
    ("Mahindra", "Thar", "Diesel", 15.2, 15.0, 2184, 4, "India", "SUV"),
    ("Honda", "City", "Petrol", 18.4, 15.0, 1498, 5, "Japan", "Sedan")
]

# List to store generated variants
new_variants = []

# Generate variants dynamically
for brand, base_model, fuel, mileage, base_price, engine, seats, country, car_type in base_variants:
    for suffix in np.random.choice(variant_suffixes, size=np.random.randint(3, 6), replace=False):
        model_name = f"{base_model} {suffix}"
        mileage_var = round(mileage + np.random.uniform(-2.0, 1.5), 2)
        price_var = round(base_price + np.random.uniform(-1.0, 2.5), 2)
        price_ex = f"{price_var - 0.5:.2f} - {price_var + 0.5:.2f} Lakh"
        transmission = np.random.choice(["Manual", "Automatic", "Manual/Automatic"])
        new_variants.append([
            brand, model_name, price_ex, fuel, mileage_var,
            transmission, engine, seats, country, price_var, car_type
        ])

# Convert new variants to DataFrame
df_variants = pd.DataFrame(new_variants, columns=df_carss.columns)

# Skip duplicates (brand + model already in df_carss)
existing_pairs = set(zip(df_carss["brand"].str.lower(), df_carss["model"].str.lower()))
filtered_rows = [
    row for row in df_variants.itertuples(index=False)
    if (row.brand.lower(), row.model.lower()) not in existing_pairs
]

# Combine with main df_carss
if filtered_rows:
    df_variants_filtered = pd.DataFrame(filtered_rows, columns=df_carss.columns)
    df_carss = pd.concat([df_carss, df_variants_filtered], ignore_index=True)

print(" Variant/Trim-level cars added successfully (duplicates skipped).")
print(" Total cars now:", len(df_carss))
print(" Total unique brands:", df_carss['brand'].nunique())
```
